Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

load the file

```python
df = pd.read_csv("car_prices.csv")
df
```

data description

```python
df.info()
```

```python
df.describe()
```

distribution of car price

```python
# Plot the distribution of car prices
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], bins=30, kde=True)
plt.title("Distribution of Car Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

most common car brand top 10

```python
# Count of car brands
top_brands = df['Make'].value_counts().head(10)
top_brands
```

chart for the most car brands

```python
# Bar plot of the most common car brands
plt.figure(figsize=(10, 5))
sns.barplot(x=top_brands.index, y=top_brands.values, palette="viridis")
plt.title("Top 10 Most Common Car Brands")
plt.xlabel("Car Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

price distribution by fuel type

```python
# Boxplot of price by fuel type
plt.figure(figsize=(10, 5))
sns.boxplot(x='fuel type', y='price', data=df)
plt.title("Car Prices by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Price")
plt.show()
```

```python
# Scatter plot for Mileage vs Price
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Mileage'], y=df['price'], alpha=0.5)
plt.title("Mileage vs Price")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.show()
```

```python
# Average price per year of manufacture
avg_price_per_year = df.groupby("Year of manufacture")["price"].mean()
avg_price_per_year
```

chrt for the average price per year

```python
# Line plot
plt.figure(figsize=(10, 5))
sns.lineplot(x=avg_price_per_year.index, y=avg_price_per_year.values)
plt.title("Average Car Price by Year of Manufacture")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.show()
```

price by use condition( nigerian use and foreign use)

```python
# Barplot of price by car condition
plt.figure(figsize=(10, 5))
sns.barplot(x='Condition', y='price', data=df)
plt.title("Car Prices by Condition")
plt.xlabel("Condition")
plt.ylabel("Price")
plt.show()
```

top 10 most expensive

```python
# Display the top 10 most expensive cars
df.nlargest(10, 'price')
```

average price by car brand top 10 most expensive

```python
# Compute average price per car brand
avg_price_brand = df.groupby("Make")["price"].mean().sort_values(ascending=False).head(10)
avg_price_brand
```

```python
# Bar plot of top 10 most expensive brands on average
plt.figure(figsize=(10, 5))
sns.barplot(x=avg_price_brand.index, y=avg_price_brand.values, palette="coolwarm")
plt.title("Top 10 Most Expensive Car Brands (Average Price)")
plt.xlabel("Car Brand")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

top 10 cars with the highest mileage

```python
# Display the top 10 cars with the highest mileage
df.nlargest(10, 'Mileage')
```

```python
# Count of cars by year
year_counts = df["Year of manufacture"].value_counts().sort_index()
year_counts
```

```python
# Line plot of car count per year
plt.figure(figsize=(10, 5))
sns.lineplot(x=year_counts.index, y=year_counts.values, marker="o")
plt.title("Number of Cars by Year of Manufacture")
plt.xlabel("Year")
plt.ylabel("Number of Cars")
plt.show()
```

price analysis by Gear type

```python
# Barplot of price distribution by gear type
plt.figure(figsize=(10, 5))
sns.barplot(x="gear type", y="price", data=df)
plt.title("Car Prices by Gear Type")
plt.xlabel("Gear Type")
plt.ylabel("Price")
plt.show()
```

```python
#most commom car colour top 10
color_counts = df["Colour"].value_counts().head(10)
color_counts
```

```python
# Bar plot of top 10 colors
plt.figure(figsize=(10, 5))
sns.barplot(x=color_counts.index, y=color_counts.values, palette="Set2")
plt.title("Top 10 Most Common Car Colors")
plt.xlabel("Color")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Barplot comparing price of "Nigerian Used" vs "Foreign Used"
plt.figure(figsize=(10, 5))
sns.barplot(x="Bought Condition", y="price", data=df)
plt.title("Car Prices by Bought Condition (Nigerian vs Imported)")
plt.xlabel("Bought Condition")
plt.ylabel("Price")
plt.show()
```
